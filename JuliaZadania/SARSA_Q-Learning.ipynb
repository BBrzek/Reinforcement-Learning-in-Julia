# Zadania SARSA i Q-Learning w języku Julia

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Tworzenie środowiska CliffWalking**

```python
import gym
env = gym.make('CliffWalking-v0')
```

**Połączenie ze środowiskiem Julia**

```python
from julia import Main
```

## SARSA
Wczytanie kodu do głównego środowiska Julia (Main)

```python
Main.include("./Julia-code/SARSA-algorithm.jl")
```

Uruchomienie algorytmu SARSA i wyświetlenie wykresu nagród.

```python
SarsaResult = Main.runSARSA(env)
plt.plot(SarsaResult)
```

Możesz uruchomić wytrenowanego agenta poprzez wywołanie w konsoli polecenia `python render.py -a SARSA`

## Q-Learning 

Nadpisanie funkcji powstałych w wyniku załadowania algorytmu SARSA na funckje Q-Learning

```python
Main.include("./Julia-code/Q-learning-algorithm.jl")
```

Uruchomienie algorytmu Q-Learning i wyświetlenie wykresu nagród.

```python
Q_learningResult = Main.runQ_learning(env)
plt.plot(Q_learningResult)
```

Możesz uruchomić wytrenowanego agenta poprzez wywołanie w konsoli polecenia `python render.py -a QLearning`

## Porównanie algorytmów SARSA i Q-Learning

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # Możesz dostosować rozmiar
ax1.plot(SarsaResult)
ax1.set_xlabel('Episodes')
ax1.set_ylabel('Rewards')
ax1.set_title('SARSA')

ax2.plot(Q_learningResult)
ax2.set_xlabel('Episodes')
ax2.set_ylabel('Rewards')
ax2.set_title('Q-learning')

plt.tight_layout()
plt.show()
```
